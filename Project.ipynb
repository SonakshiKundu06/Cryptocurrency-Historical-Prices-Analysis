```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
dataset = pd.read_csv("C:\\Users\\insp 5430\\programmings\\Data Visualization\\coin_Aave.csv")
df = dataset.copy()
df['Date'] = pd.to_datetime(df['Date'])
df['TimeIndex'] = df['Date'].map(pd.Timestamp.toordinal)

X = df[['TimeIndex', 'Volume']]
y = df['Close']

model = LinearRegression()
model.fit(X, y)

df['Predicted_Close'] = model.predict(X)
plt.figure(figsize=(12,6))

ax = plt.gca()
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))  # Show every 3 months
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45, ha='right')

plt.plot(df['Date'], df['Close'], label='Actual')
plt.plot(df['Date'], df['Predicted_Close'], label='Predicted')
plt.legend()
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Price Prediction Model")
```
